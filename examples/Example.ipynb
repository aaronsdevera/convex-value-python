# convex-value-python example

Import the library.

```python
from convexvalue import ConvexValue
```

Create an instance of the class object. It'll check for credentials stored as the environment variables `CONVEX_VALUE_EMAIL` and `CONVEX_VALUE_PASSWORD` in order to perform the login.

```python
cvx = ConvexValue()
```

## Querying with SQL

```python
SQL_QUERY = '''select
  *
  from opt_summary
  where symbol like '.SPX%'
  limit 5
'''
```

Execute query.

```python
results = cvx.query_sql(SQL_QUERY)
```

Display results as a `pandas` DataFrame.

```python
results.df()
```

Can plot as well.

```python
results.plot()
```
